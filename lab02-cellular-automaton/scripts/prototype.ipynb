```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
times = ["0", "1", "2", "5", "10", "20", "50", "100"]

names = ["../output/evolution_run_t" + t + ".csv" for t in times]

dfs = [pd.read_csv(name, index_col=0) for name in names]
```

```python
ig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes_flat = axes.flatten()

for i, df in enumerate(dfs):
    ax = axes_flat[i]
    sns.heatmap(df, cmap='viridis', ax=ax, cbar=False, xticklabels=False, yticklabels=False)
    ax.set_title(f"t = {times[i]}")

plt.tight_layout()
plt.show()
```
